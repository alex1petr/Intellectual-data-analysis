```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="joeddav/xlm-roberta-large-xnli")
```

```python
text = "Цей фільм мені взагалі не сподобався."
candidate_labels = ["позитивний", "негативний", "нейтральний"]

result = classifier(text, candidate_labels)
print(result)
```

```python
from transformers import pipeline
summarizer = pipeline("summarization", model="google/mt5-small")
```

```python
text = """
summarize: Україна активно готується до впровадження масштабних реформ у сфері освіти, спрямованих на підвищення якості навчання та модернізацію освітніх процесів у школах і університетах. Нові ініціативи передбачають перегляд навчальних програм, оновлення методик викладання та інтеграцію сучасних технологій у освітній процес. Зокрема, планується посилення практичної спрямованості навчання, розвиток критичного мислення та навичок вирішення проблем у студентів і школярів. Окрім цього, реформи мають на меті створення більш гнучкої та інклюзивної системи освіти, яка відповідала б міжнародним стандартам і потребам сучасного ринку праці. Очікується, що такі зміни сприятимуть не лише професійному розвитку молоді, а й загальному підвищенню рівня освіченості населення та конкурентоспроможності країни на глобальному рівні.
"""
summary = summarizer(text, do_sample=False)
print(summary[0]['summary_text'])
```

```python
from transformers import pipeline
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-uk-en")
```

```python
text = "Я люблю програмувати."
translation = translator(text)
print(translation[0]['translation_text'])
```

```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipe = pipe.to("cuda")
```

```python
from IPython.display import display
prompt = "Red car"
image = pipe(prompt).images[0]
display(image)
image.save("prompt_img.png")
```
